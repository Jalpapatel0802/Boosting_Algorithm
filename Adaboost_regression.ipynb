```python
filterdata=df.drop(['id', 'date', 'price', 'sqft_lot', 'floors', 'condition', 'yr_renovated', 'zipcode'],axis=1)
```

```python
filterdata.dtypes
```

```python
X=filterdata
y=df.price.values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.2,random_state=100)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
adaboost=AdaBoostRegressor(n_estimators=50,learning_rate=0.2,loss='exponential').fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
predict=adaboost.predict(X_test)
r2score = r2_score(predict,y_test)
```

```python
r2score 
```

```python
from sklearn.metrics import explained_variance_score,mean_squared_error
import math
```

```python
# Calculating Mean Sqaured Error & Root Mean Squared Error
mse = mean_squared_error(y_test, predict)

rmse = math.sqrt(mse)

print(rmse)
```
